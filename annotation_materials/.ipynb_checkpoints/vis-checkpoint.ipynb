```python
import open3d as o3d
import numpy as np

model_mesh = o3d.io.read_triangle_mesh('cube.obj')
pts_num = np.asarray(model_mesh.vertices).shape[0]
model_mesh.vertex_colors = o3d.utility.Vector3dVector(np.array([[0, 0.651, 0.929]]).repeat(pts_num, axis=0))

axis_pcd = o3d.geometry.TriangleMesh.create_coordinate_frame(size=0.0001, origin=[0, 0, 0])

o3d.visualization.draw_geometries([model_mesh])
```
